```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as AA
```

```python
data = np.genfromtxt("IgnSolution.dat", dtype=str)
Header = (data[0,:]).tolist()
solTchem = (data[1:,:]).astype(np.float)
```

```python
print(np.shape(solTchem))
```

```python
x = solTchem[:,1]
sp=50
for var in Header[:10]:
    plt.figure()
    plt.plot(x,solTchem[:,Header.index(var)])
    plt.plot(x[sp:sp+2],solTchem[sp:sp+2,Header.index(var)],'r')
    plt.xlabel('t [s]',fontsize=16)
    plt.ylabel(var,fontsize=16)
```

```python
density = solTchem[sp,Header.index('Density[kg/m3]')]
pressure = solTchem[sp,Header.index('Pressure[Pascal]')]
mass_fraction_and_temp= solTchem[sp,5:]
state =  np.hstack((np.atleast_1d(density),np.atleast_1d(pressure),np.atleast_1d(mass_fraction_and_temp)))
```

```python
np.savetxt('inputGas.dat',state)
```
